```python
import numpy as np
data = np.load('data/S001_S001R03.edf.npy')
print(data)
```

```python
hands = [3, 4, 7, 8, 11, 12]
```

```python
allData = None

for i in range(3, 14):
    # make i a string and 2 digits

    if i not in hands:
        print(f'Skipping {i}')
        continue

    i = str(i).zfill(2)
    data = np.load(f'data/S001_S001R{i}.edf.npy')
    if allData is None:
        allData = data
    else:
        allData = np.vstack((allData, data))
```

```python
allData.shape
```

```python
df = pd.DataFrame(allData)
df
```

```python
t0 = df[df[1] == "T0"]
t1 = df[df[1] == "T1"]
t2 = df[df[1] == "T2"]

minSize = min(t0.shape[0], t1.shape[0], t2.shape[0])

t0Sample = t0.sample(n=minSize)
t1Sample = t1.sample(n=minSize)
t2Sample = t2.sample(n=minSize)

dfSample = pd.concat([t0Sample, t1Sample, t2Sample])
dfSample
```

```python
# save to .npy file
np.save('cleanedData.npy', dfSample.values)
```
